```python
#Analysoidaan Seligson Top 25 Global Brands rahaston arvostustasoa. Käytetään tähän P/E-lukua. P/E luku on tietenkin vain yksi
#arvostuksen mittari, mutta se toimii parhaiten samanlaisten ja saman sektorien yhtiöiden kanssa. Rahastossa on tässä suhteessa
#riittävän vertailukelpoisia osakkeita. Otetaan vertailuun kaikki osakkeet, joita on yli 3% rahaston omistuksista.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Historiallisia P/E lukuja ei ole saatavilla mistään valmiista kirjastosta, joten ne on laitettava käsin. P/E luvut alkavat
#vuodesta 2015 ja on otettu jokaisen vuoden 31.12 päivä.
```

```python
alphabet = [28.42, 58.46, 23.88, 27.21, 29.87, 25.79, 19.33, 24.06]
pg = [12.37, 20.51, 19.29, 63.08, 23.96, 26.94, 25.45, 24.09]
cocacola =  [21.93, 21.92, 26.59, 23.24, 27.53, 24.40, 27.73, 23.50]
apple = [12.81, 16.30, 12.37, 22.47, 35.05, 29.04, 21.88, 29.88]
pepsi = [19.32, 29.36, 10.76, 23.13, 26.25, 29.51, 24.96, 25.50]
prada = [29.20, 21.00, 35.00, 35.20, 36.70, 27.30, 24.60, 24.30]
amazon = [152.30, 189.70, 74.50, 80.11, 77.73, 51.33, 100.67, 52.39]
colgate = [20.02, 28.16, 19.87, 22.38, 24.93, 31.33, 35.47, 28.30]
aexpress = [11.78, 30.10, 11.08, 14.63, 30.45, 15.68, 14.59, 16.56]
unilever = [21.80, 21.80, 16.00, 18.10, 22.50, 20.9, 17.7, 14.80]
disney = [17.98, 14.83, 14.82, 24.35, 19.51, 47.40, 47.40, 55.40]
adidas = [28.90, 24.90, 21.80, 31.10, 117.10, 32.90, 30.00, 38.40]
kellanova = [36.30, 30.00, 10.8, 32.40, 18.00, 17.40, 16.30, 41.40]
yum = [13.81, 19.26, 17.74, 22.39, 34.40, 25.40, 27.40, 23.14]
mondelez = [36.41, 19.98, 15.87, 18.48, 21.84, 20.50, 32.83, 19.74]
danone = [21.60, 20.60, 15.90, 23.50, 18.20, 18.30, 24.50, 34.60]
```

```python
#Nyt on käsin kerätty 2016-2023 P/E luvut. Seuraavaksi visualisoidaan data, mutta sitä ennen muutetaan P/E lukujen muutos
#prosentuaalisiksi. Aloitetaan luomalla DataFrame.
```

```python
df = pd.DataFrame({'Alphabet': alphabet, 'pg': pg, 'cocacola': cocacola, 'apple': apple, 'pepsi':pepsi, 'prada': prada,
                  'amazon': amazon, 'colgate': colgate, 'aexpress': aexpress, 'unilever': unilever, 'disney': disney,
                  'adidas': adidas, 'kellanova': kellanova, 'yum': yum, 'mondelez': mondelez, 'danone': danone})
df_pct = df.pct_change().iloc[1:] #Muutetaan prosenteiksi, josta vertailu aloitetaan ensimmäisestä ja poistetaan ensimmäinen,
#koska sillä ei ole aikaisempia vertailukohteita
```

```python
print(df_pct)
```

```python
#Nyt data on oikeassa muodossa ja aloitetaan visualisointi.
```

```python
vuodet = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
plt.figure(figsize=(14,7))

plt.plot(vuodet, df_pct['Alphabet']*100, label= 'Alphabet')
plt.plot(vuodet, df_pct['pg']*100, label= 'Procter & Gamble')
plt.plot(vuodet, df_pct['cocacola']*100, label='Coca Cola')
plt.plot(vuodet, df_pct['apple']*100, label='Apple')
plt.plot(vuodet, df_pct['pepsi']*100, label='Pepsi')
plt.plot(vuodet, df_pct['prada']*100, label='Prada')
plt.plot(vuodet, df_pct['amazon']*100, label='Amazon')
plt.plot(vuodet, df_pct['colgate']*100, label='Colgate')
plt.plot(vuodet, df_pct['aexpress']*100, label='American Express')
plt.plot(vuodet, df_pct['unilever']*100, label='Unilever')
plt.plot(vuodet, df_pct['disney']*100, label='Disney')
plt.plot(vuodet, df_pct['adidas']*100, label='Adidas')
plt.plot(vuodet, df_pct['kellanova']*100, label='Kellanova')
plt.plot(vuodet, df_pct['yum']*100, label='Yum Brands')
plt.plot(vuodet, df_pct['mondelez']*100, label = 'Mondelez')
plt.plot(vuodet, df_pct['danone']*100, label= 'Danone')

ax = plt.gca()
ax.set_facecolor('black')
plt.xlabel('Vuodet')
plt.ylabel('Muutos (%) P/E luvuissa')

plt.legend()
```

```python
#Nähdään, että P/E luvut muutamia hetkellisiä poikkeuksia lukuun ottamatta ovat luvut pitkälti pysyneet samanlaisina vuosien
#saatossa. Tämä ei tule mitenkään yllätyksenä, koska yhtiöt ovat jo erittäin kypsän vaiheen yhtiöitä niin mitään kasvuyhtiöiden
#P/E lukuja ei ole odotettavissakaan.
```

```python
#Selvitetään vielä muutamia kiinnostavia tietoja 1. Kunkin yrityksen P/E lukujen mediaani, 2. Milloin pienin, 3. Mikä
#vaihteluväli
```

```python
mediaanit = df.median() #Mediaani
suurin_mediaani = mediaanit.max() #Suurin mediaani
pienin_mediaani = mediaanit.min() #Pienin mediaani
vaihteluvali = suurin_mediaani - pienin_mediaani
```

```python
print(f"Mediaanit: \n{mediaanit}")
```

```python
print(f"Suurin mediaani: {suurin_mediaani}")
print(f"Pienin mediaani: {pienin_mediaani}")
print(f"Vaihteluväli: {vaihteluvali}")
```
